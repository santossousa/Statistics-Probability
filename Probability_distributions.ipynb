Para entender o conceito de distribuições de probabilidade, é importante começar com variáveis aleatórias. Eles são usados para modelar distribuições de probabilidade em análises estatísticas.

Uma variável aleatória é um resultado numérico de algum processo aleatório ou uma função que atribui valores a cada resultado de um experimento. Geralmente é indicado pelo símbolo X.

Variáveis aleatórias são divididas em duas categorias:

1. **Variáveis aleatórias discretas** podem assumir um número finito e contável de valores. Por exemplo, o resultado de uma jogada de dados pode ser 1, 2, 3, 4, 5 ou 6.

2. **Variáveis aleatórias contínuas** podem assumir um número infinito de valores. Exemplos incluem medidas como temperatura, altura e peso.

A **função de massa de probabilidade (PMF)** de uma variável aleatória discreta é uma função que fornece a probabilidade de que a variável assuma um determinado valor discreto. Mas para uma variável contínua, não podemos encontrar a probabilidade absoluta. Por que? Como vemos, com variáveis contínuas, o número de resultados possíveis é infinito.


Considere o peso como um exemplo. O peso de uma pessoa pode assumir uma infinidade de valores, como 25,0001 kgs, 25,0000003 kgs e assim por diante. Assim, não é possível calcular a probabilidade absoluta de o peso de uma pessoa ser exatamente 25 kg, o que resultaria em zero.

Para superar essa limitação, usamos a **função de densidade de probabilidade** ou **PDF** para variáveis contínuas, que serve ao mesmo propósito que a função de massa de probabilidade ou PMF para variáveis discretas. O PDF representa a probabilidade de que o valor de uma variável aleatória contínua caia dentro de um determinado intervalo de valores.

A **função de distribuição cumulativa** ou **CDF** fornece a probabilidade de uma variável aleatória ser menor ou igual a um valor especificado. É calculado como a integral da PDF e representa a área sob a curva da PDF até certo ponto.

Os tipos típicos de distribuições de probabilidade para variáveis aleatórias discretas incluem *Binomial*, *Uniform* e *Poisson*.

## Distribuição binomial
Uma distribuição binomial descreve a probabilidade de obter um número especificado de sucessos em um determinado número de tentativas de um experimento binomial.

Um experimento binomial é um experimento que segue estas propriedades:

1. O experimento consiste em $n$ tentativas repetidas.
2. Cada tentativa tem dois resultados possíveis.
3. A probabilidade de sucesso ($p$) e falha ($1-p$) é a mesma para cada tentativa.
4. Cada tentativa é independente uma da outra.
Por exemplo, considere lançar uma moeda não viciada $n$ vezes. Neste exemplo, a probabilidade de que o resultado seja cara pode ser considerada igual a $p$, enquanto $1-p$ é a probabilidade de coroa. Cada vez que a moeda é lançada, o resultado é independente de todas as outras tentativas.

Se uma variável aleatória $X$ segue uma distribuição binomial, a probabilidade de que $X$ seja igual a $k$ sucessos pode ser encontrada pela seguinte fórmula:
  
$$P(X=k) = ^n C_k p^k (1-p)^{n-k}$$
  
onde:
  
$p$ é a probabilidade de sucesso, $(1-p)$ é a probabilidade de falha
$n$ é o número de tentativas

A distribuição binomial tem as seguintes características:
**Média** = $n*p$ (número de tentativas * probabilidade de sucesso)
**Variação** = $npq$ (número de tentativas * probabilidade de sucesso * probabilidade de falha)


**Exemplo:**

Suponha que, de acordo com os últimos relatórios policiais, $80\%$ de todos os pequenos crimes não sejam resolvidos e, em sua cidade, pelo menos três desses pequenos crimes sejam cometidos. Os três crimes são independentes entre si. A partir dos dados fornecidos, qual é a probabilidade de que um dos três crimes seja resolvido?

solução: **Calculando Probabilidades Binomiais**

$$P(X = 1) = ^{3}C_1(0.2)^1(0.8)^{2} = 0.384$$


```python
import scipy.stats as stats
n, r, p = 3, 1, 0.2
print(f'{stats.binom.pmf(r, n, p):.3f}')
```

### **Efeito de n e p na Forma**

1. Para *p* pequeno e *n* pequeno, a distribuição binomial é o que chamamos de enviesada para a direita.
2. Para *p* grande e *n* pequeno, a distribuição binomial é o que chamamos de enviesada à esquerda.
3. Para *p*=0,5 e grande e pequeno *n* , a distribuição binomial é o que chamamos de simétrica.
4. Para *p* pequeno e *n* grande, a distribuição binomial se aproxima da simetria.

```python
from numpy import random
import matplotlib.pyplot as plt 
import seaborn as sns 


plt.figure(figsize=(8,6))
sns.set_style("whitegrid")

sample1 = random.binomial(n = 15, p = 0.2, size = 250)
sample2 = random.binomial(n = 15, p = 0.8, size = 250)
sample3 = random.binomial(n = 15, p = 0.5, size = 250)
sample4 = random.binomial(n = 40, p = 0.2, size = 250)

sns.kdeplot(sample1, label="sample1")
sns.kdeplot(sample2, label="sample2")
sns.kdeplot(sample3, label="sample3")
sns.kdeplot(sample4, label="sample4")

plt.legend(labels=["sample1","sample2","sample3","sample4"])

plt.show()
```

## Distribuição geométrica

A distribuição geométrica descreve a probabilidade de experimentar uma certa quantidade de falhas antes de experimentar o primeiro sucesso em uma série de ***tentativas*** de Bernoulli. Um ensaio de Bernoulli é um experimento com apenas dois resultados possíveis – “sucesso” ou “fracasso” – e a probabilidade de sucesso é a mesma sempre que o experimento é conduzido. Um exemplo de tentativa de Bernoulli é o cara ou coroa. A moeda só pode cair em dois lados (poderíamos chamar cara de “sucesso” e coroa de “fracasso”) e a probabilidade de sucesso em cada lançamento é de 0,5, assumindo que a moeda é honesta.

Se uma variável aleatória X segue uma distribuição geométrica, então a probabilidade de experimentar k falhas antes de experimentar o primeiro sucesso pode ser encontrada pela seguinte fórmula:

$$P(X=k) = (1-p)^kp$$

onde:

$k$ é o número de falhas antes do primeiro sucesso
$p$ é a probabilidade de sucesso em cada tentativa

**Exemplo**:

 Suponha que queremos saber quantas vezes teremos que jogar uma moeda honesta até que dê cara. Podemos usar a fórmula acima para determinar a probabilidade de ocorrer 0, 1, 2, 3 falhas, etc. antes que a moeda dê cara:

**Observação:** A moeda pode apresentar 0 'falha' se der cara no primeiro lançamento.

$P(X=0) = (1-.5)^0(.5) = 0,5$

$P(X=1) = (1-.5)^1(.5) = 0,25$

$P(X=2) = (1-.5)^2(.5) = 0,125$

$P(X=3) = (1-.5)^3(.5) = 0,0625$

## Distribuição uniforme

A distribuição uniforme é uma distribuição de probabilidade na qual cada valor entre um intervalo de $a$ a $b$ é igualmente provável de ocorrer.

Se uma variável aleatória $X$ segue uma distribuição uniforme, então a probabilidade de $X$ assumir um valor entre $x_1$ e $x_2$ pode ser encontrada pela seguinte fórmula:

$$P(x_1 < X < x_2) = \frac{(x_2 – x_1)}{(b – a)}$$

onde:

$x_1$: o menor valor de juros
$x_2$: o valor superior dos juros
$a$: o valor mínimo possível
$b$: o valor máximo possível
**negrito**

**Exemplo**:

Suponha que o peso dos golfinhos seja distribuído uniformemente entre 100 libras e 150 libras.

Se selecionarmos aleatoriamente um golfinho ao acaso, podemos usar a fórmula acima para determinar a probabilidade de o golfinho escolhido pesar entre 120 e 130 libras:

$P(120 < X < 130) = (130 – 120) / (150 – 100) = 10 / 50 = 0,2$


A probabilidade de o golfinho escolhido pesar entre 120 e 130 libras é de 0,2.

**Propriedades da Distribuição Uniforme**

A distribuição uniforme tem as seguintes propriedades:

+ Média: (a + b) / 2
+ Mediana: (a + b) / 2
+ Desvio padrão: √(b – a)2/12
+ Variância: (b – a)2/12

```python
dist_uniforme = stats.uniform(loc=100, scale=(150-100))

print(f'{dist_uniforme.cdf(130) - dist_uniforme.cdf(120):.2f}')
```

## **Distribuição de Poisson**


Novamente, para entender a distribuição de Poisson, primeiro precisamos entender o que são os **experimentos de Poisson**.

Um experimento de Poisson é um experimento que possui as seguintes propriedades:

+ O número de sucessos no experimento pode ser contado.
+ O número médio de sucessos que ocorre durante um intervalo de tempo (ou espaço) específico é conhecido.
+ Cada resultado é independente.
+ A probabilidade de ocorrer um sucesso é proporcional ao tamanho do intervalo


Um exemplo de experimento de Poisson é o número de nascimentos por hora em um determinado hospital. Por exemplo, suponha que um determinado hospital tenha uma média de 10 partos por hora. Este é um experimento de Poisson porque tem as seguintes quatro propriedades:

+ O número de sucessos no experimento pode ser contado – Podemos contar o número de nascimentos.
+ É conhecido o número médio de sucessos que ocorrem durante um intervalo de tempo específico – Sabe-se que ocorrem em média 10 nascimentos por hora.
+ Cada resultado é independente – A probabilidade de uma mãe dar à luz durante uma determinada hora é independente da probabilidade de outra mãe dar à luz.
+ A probabilidade de ocorrer um sucesso é proporcional ao tamanho do intervalo – quanto maior o intervalo de tempo, maior a probabilidade de ocorrer um nascimento.

Podemos usar a distribuição de Poisson para responder a perguntas sobre probabilidades relacionadas a esse experimento de Poisson, como:

+ Qual é a probabilidade de ocorrerem mais de 12 nascimentos em uma determinada hora?
+ Qual é a probabilidade de ocorrerem menos de 5 nascimentos em uma determinada hora?
+ Qual é a probabilidade de ocorrerem entre 8 e 11 nascimentos em uma determinada hora?

Se uma variável aleatória $X$ segue uma distribuição de Poisson, então a probabilidade de que $X = k$ sucessos pode ser encontrada pela seguinte fórmula:

$$P(x=k)=\frac{\lambda^k e^{-\lambda}}{k!}$$

onde $P(x=k)$ é a probabilidade do evento ocorrer $k$ número de vezes, $k$ é o número de ocorrências do evento e $\lambda$ representa o número médio de eventos que ocorrem durante um intervalo específico.

A distribuição de Poisson pode ser usada para calcular o número de ocorrências que ocorrem em um determinado período, por exemplo:
+ número de chegadas a um restaurante por hora
+ número de acidentes de trabalho ocorridos em uma fábrica ao longo de um ano
+ número de reclamações de clientes em call center em uma semana

Propriedades de uma distribuição de Poisson:
1. Média=variância=$\lambda$. Em uma distribuição de Poisson, a média e a variância têm os mesmos valores numéricos.
2. Os eventos são independentes, aleatórios e não podem ocorrer ao mesmo tempo.


O eixo horizontal é o índice k, o número de ocorrências. λ é a taxa esperada de ocorrências. O eixo vertical é a probabilidade de k ocorrências dado λ. A função é definida apenas em valores inteiros de k; as linhas de conexão são apenas guias para os olhos.

**Exemplo:**

Em uma estação de metrô, a média de máquinas automáticas fora de operação é de duas. Supondo que o número de máquinas fora de operação segue uma distribuição de Poisson, calcule a probabilidade de que um determinado ponto no tempo:
1. Exatamente três máquinas estão fora de operação
2. Mais de duas máquinas estão fora de operação

```python


print(f'Exatamente três máquinas estão fora de operação {stats.poisson.pmf(3,2):.2f}')
print(f'Mais de duas máquinas estão fora de operação {1 - stats.poisson.cdf(2,2):.2f}')
```

## **Distribuição normal**
Uma distribuição normal é uma curva simétrica em forma de sino, definida por sua média ($\mu$) e desvio padrão ($\sigma$)

Características de uma distribuição normal:
1. O valor central ($\mu$) também é a moda e a mediana para uma distribuição normal
2. Verificando a normalidade: Em uma distribuição normal, a diferença entre o valor do percentil 75 ($Q_3$) e o valor do percentil 50 (mediana ou $Q_2$) é igual à diferença entre a mediana ($Q_2$) e o percentil 25 ($Q_1$). Em outras palavras,

$$Q_3 - Q_2 = Q_2 - Q_1$$


Se a distribuição for assimétrica, esta equação não se sustenta.

+ Em uma distribuição assimétrica à direita, $(Q_3 − Q_2)> (Q_2 - Q_1)$
+ Em uma distribuição assimétrica à esquerda, $(Q_2 - Q_1) > (Q_3 - Q_2)$


 **Exemplo**:
 
  A concentração de um poluente em água liberada por uma fábrica tem distribuição N(8,1.5). Qual a chance, de que num dado dia, a concentração do poluente exceda o limite regulatório de 10 ppm?

A solução do problema resume-se em determinar a proporção da distribuição que está acima de 10 ppm, ie $P(X>10)$. Usando a estatística $Z$ temos:

 $$P(X>10)=P(Z>\frac{10-8}{1.5})=P(Z>1.33)=1-P(Z \leq 1.33)=0.09$$

```python
print(f'Espera-se que a água liberada pela fábrica exceda os limites regulatórios cerca de {(1 - stats.norm.cdf(10, loc=8, scale=1.5))*100:.2f}%')
```

## **Distribuição normal padrão**


Para padronizar unidades e comparar distribuições com diferentes médias e variâncias, usamos uma distribuição normal padrão.

Propriedades de uma distribuição normal padrão:
+ A distribuição normal padrão é uma distribuição normal com um valor médio de 0 e um desvio padrão de 1.
+ Qualquer distribuição normal pode ser convertida em distribuição normal padrão usando a seguinte fórmula:

$$z = \frac {x-\mu}{\sigma}$$

onde $\mu$ e $\sigma$ são a média e a variância do
distribuição normal original.

A **pontuação z** (também chamada de **pontuação padrão**) dá uma ideia de quão longe da média um ponto de dados está.

Em uma distribuição normal padrão,
+ 68,2% dos valores estão dentro de 1 desvio padrão da média
+ 95,4% dos valores situam-se entre 2 desvios padrão da média
+ 99,8% estão dentro de 3 desvios padrão da média
+ A área sob a distribuição normal padrão entre quaisquer dois pontos representa a proporção de valores que se encontra entre esses dois pontos. Por exemplo, a área sob a curva em ambos os lados da média é 0,5. Dito de outra forma, 50% dos valores estão em ambos os lados da média.

**Exemplo**:



Suponha que você esteja estudando o desempenho de alunos em um exame. A pontuação média do exame é 70 e o desvio padrão é 10. Você deseja calcular a probabilidade de um aluno ser classificado como "acima da média", ou seja, obter uma pontuação maior que 70.

```python
  print(f'A probabilidade de obter uma pontuação maior que 80 é de aproximadamente  {1 - stats.norm.cdf(80,70,10):.4f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(30, 100, 1000)

y = stats.norm.cdf(x, loc=70, scale=10)

plt.plot(x, y, 'b-', label='Distribuição Normal')

x_highlight = np.linspace(70.1, 100, 100)
y_highlight = 1 - stats.norm.cdf(x_highlight, loc=70, scale=10)
plt.fill_between(x_highlight, y_highlight, color='orange', alpha=0.5, label='Acima da média')

plt.xlabel('Pontuação')
plt.ylabel('Probabilidade Acumulada')
plt.title('Distribuição Normal - Probabilidade Acima da Média')
plt.legend()

plt.show()
```
